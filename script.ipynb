```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a SQL connection to our SQLite database
dbfile = "data\movie.sqlite"
con = sqlite3.connect(dbfile)
cur = con.cursor()
```

```python
# query imdb table headers
imdb_query_headers = [a for a in cur.execute("PRAGMA table_info(IMDB)")]
imdb_table_column_headers = [i[1] for i in imdb_query_headers]
# query imdb table data
imdb_table_data = [a for a in cur.execute("SELECT * FROM IMDB")]
#Save imdb table into dataframe
imdb_df = pd.DataFrame(imdb_table_data, columns=imdb_table_column_headers)

#DATA PROCESSING
#Change data type of some columns
imdb_df[["Budget", "MetaCritic"]] = imdb_df[["Budget", "MetaCritic"]].apply(pd.to_numeric)
#Split number and "min" in Runtime column, and change type
imdb_df["Runtime"] = imdb_df["Runtime"].str.split(" ", expand=True)[0]
imdb_df["Runtime"] = imdb_df["Runtime"].apply(pd.to_numeric)
```

```python
# query EARNING table headers
earning_query_headers = [a for a in cur.execute("PRAGMA table_info(earning)")]
earning_headers = [i[1] for i in earning_query_headers]
# query earning table data
earning_table_data = [a for a in cur.execute("SELECT * FROM earning")]
# Save earning table into dataframe
earning_df = pd.DataFrame(earning_table_data, columns=earning_headers)
```

```python
# query GENRE table headers
genre_query_headers = [a for a in cur.execute("PRAGMA table_info(genre)")]
genre_headers = [i[1] for i in genre_query_headers]
# query genre table data
genre_table_data = [a for a in cur.execute("SELECT * FROM genre")]
# Save genre table into dataframe
genre_df = pd.DataFrame(genre_table_data, columns=genre_headers)
```

```python
# Be sure to close the connection
con.close()
```
